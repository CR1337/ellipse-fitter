```python
from polyline_factory import PolylineFactory
import matplotlib.pyplot as plt
```

```python
xc = 0
yc = 0
a =10
b = 20
theta = 0
n_points = 100

n_test_points = 50 

factory = PolylineFactory(xc, yc, a, b, theta, n_points)
polyline = factory.buildPolyline()
inside_points = factory.generateInsidePoints(n_test_points)
outside_points = factory.generateOutsidePoints(n_test_points)
```

```python
for p1, p2 in zip(polyline[:-1], polyline[1:]):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b')
for p in inside_points:
    plt.plot(p[0], p[1], 'go')
for p in outside_points:
    plt.plot(p[0], p[1], 'ro')
```
